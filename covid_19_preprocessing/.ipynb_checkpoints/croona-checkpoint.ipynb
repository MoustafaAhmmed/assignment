## handelling recovered/death data

```python
df["death"].value_counts()
```

```python
df["death(y_n)"]=df["death"].apply(lambda x: x if x=="1" or x=="0" else "1")
```

```python
df["death(y_n)"].value_counts()
```

```python
sns.countplot(df["gender"],hue=df["death(y_n)"])
```

```python
df["recovered(y_n)"]=df["recovered"].apply(lambda x: x if x=="1" or x=="0" else "1")
```

```python
sns.countplot(df["gender"],hue=df["recovered(y_n)"])
```

```python
sns.swarmplot(y=df["age"],x=df["recovered(y_n)"],hue=df["gender"])
```

```python
df["period"]=df["exposure_end"]-df["exposure_start"]
df["period"].mean()
```

## missing data

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="most_frequent")
df["symptom"] = imputer.fit_transform(df[["symptom"]])
```

```python
df[df["case_in_country"].isnull()].tail(3)
```

```python
df[df["country"]=="China"]["case_in_country"].fillna(df["id"],inplace=True)
```
